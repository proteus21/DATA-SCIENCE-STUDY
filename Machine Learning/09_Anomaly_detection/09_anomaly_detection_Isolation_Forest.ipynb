### <a name='0'></a> Import libraries

````python
# Import the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

sns.set(font_scale=1.2)
np.random.seed(10)
````

### <a name='1'></a> Data generation

````python
data = pd.read_csv('https://storage.googleapis.com/esmartdata-courses-files/ml-course/factory.csv')
data.head()
````

````python
data.describe()
````

### <a name='2'></a> Data visualisation

````python
fig = go.Figure()
fig=px.scatter(data, x='item_length', y='item_width', width=950, template='plotly_dark', title='Isolation Forest')
fig.show()
````

### <a name='3'></a> Isolation Forest

````python
from sklearn.ensemble import IsolationForest

outlier=IsolationForest(n_estimators=100, contamination=0.05)
outlier.fit(data)
````

````python
y_pred=outlier.predict(data)
y_pred[:30]
````

### <a name='4'></a> Isolation Forest - visualisation

````python
fig=go.Figure()
fig=px.scatter(data, x='item_length', y='item_width', color='item_length', width=950, template='plotly_dark',color_continuous_midpoint=-1, title='Isolation Forest')
fig.show()
````
