```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
np.random.seed(42) # ziarno losowe
np.set_printoptions(edgeitems=30, linewidth=100000, 
                    formatter=dict(float=lambda x: f'{x:.2f}')) #aby wiecej sobie podejrzec

sns.set()
sklearn.__version__
```

###INPUT DATA / Załadowanie   

```python
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

Preapare data /Przygotowanie danych

```python
data=raw_data_copy['data']
target=raw_data_copy ['target']
print(f'{data[:5]}\n')
print(target[:5])  # na dole klasa ze zmienną docelowa
```

###join atributes with target vairable / połaczenie atrybutów ze zmienną docelowa

```python
all_data=np.c_[data,target]
all_data[:5]
```

###Build Data Frame object / budowa obiektu data frame


```python
df=pd.DataFrame(data=all_data,columns=raw_data.feature_names+['target'])
df.head()
```

```python
data=df.copy()
target=data.pop('target') # wyodrebnienie. Pop wydrebnia wszystkie kolunny bez zaznaczonej
data.head()
```

```python
target.head()
```

###Split data on two sets: training and test / Podzial danych na zbior treningowy i testowy - iris data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data, target, random_state=42)

print(f'X_train shape{ X_train.shape}')
print(f'y_train shape{ y_train.shape}')
print(f'X_test shape{ X_test.shape}')
print(f'y_test shape{ y_test.shape}')
print(f'\nTest ratio{len(X_test)/len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
X_train, X_test, y_train, y_test=train_test_split(data, target, random_state=42, test_size=0.3) # test_size  to 30%
print(f'X_train shape{ X_train.shape}')
print(f'y_train shape{ y_train.shape}')
print(f'X_test shape{ X_test.shape}')
print(f'y_test shape{ y_test.shape}')
print(f'\nTest ratio{len(X_test)/len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
X_train, X_test, y_train, y_test=train_test_split(data, target, random_state=42, test_size=0.3,stratify=target ) # test_size  to 30%, stratify - podział na rowne 
print(f'X_train shape{ X_train.shape}')
print(f'y_train shape{ y_train.shape}')
print(f'X_test shape{ X_test.shape}')
print(f'y_test shape{ y_test.shape}')
print(f'\nTest ratio{len(X_test)/len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

###Split data on two sets: training and test / Podzial danych na zbior treningowy i testowy - breast cancer data

```python
from sklearn.datasets import load_breast_cancer

raw_data = load_breast_cancer()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
data=raw_data_copy['data']
target=raw_data_copy ['target']
print(f'{data[:5]}\n')
print(target[:5])  # na dole klasa ze zmienną docelowa
```

Do równego podzialu ponizej

```python
data=df.copy()
target=data.pop('target')
data.head()
```

test_size to rozmiar testowy zbioru

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data, target, random_state=40, test_size=0.25)

print(f'X_train shape{ X_train.shape}')
print(f'y_train shape{ y_train.shape}')
print(f'X_test shape{ X_test.shape}')
print(f'y_test shape{ y_test.shape}')
print(f'\nTest ratio{len(X_test)/len(data):.2f}')
print(f'\ntarget:\n{target.value_counts()/len(target)}')
print(f'\ny_train:\n{y_train.value_counts()/len(y_train)}')
print(f'\ny_test:\n{y_test.value_counts()/len(y_test)}')
```

###Equal split in relative  to target variable /Rowny podział ze wzgledu na zmienną docelowa

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=40, test_size=0.25, stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ntarget:\n{target.value_counts() / len(target)}')
print(f'\ny_train:\n{y_train.value_counts() / len(y_train)}')
print(f'\ny_test:\n{y_test.value_counts() / len(y_test)}')
     
```
