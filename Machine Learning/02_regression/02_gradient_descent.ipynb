### <a name='0'></a> Import bibliotek  /Import library

```python
import numpy as np
import pandas as pd
import plotly.express as px
np.random.seed(42)
```

### <a name='1'></a> Przygotowanie danych /Prepare input data

```python
X1=np.array([1,2,3,4,5,6])
Y=np.array([3000,3250,3500,3750,4000,4250])
m=len(X1)


print(f'Lata pracy / Years of the work {X1}')
print(f'Wynagrodzenie / Solary {Y}')
print(f'Liczba próbek / Number of trials {m}')
```

```python
X1=X1.reshape(m,1) #
Y=Y.reshape(-1,1)
print(X1)
print(X1.shape)
```

```python
bias=np.ones((m,1)) # wyraz wolny / Matrix with value 1
print(bias)
print(bias.shape)
```

```python
X=np.append(bias,X1,axis=1)
print(X)
print(X.shape)
```

### <a name='4'></a> Losowa inicjalizacja parametrów / Random parameter initialization

```python
eta=0.01 # wskaznik uczenia / learn factor
weights=np.random.randn(2,1) #Return a sample (or samples) from the “standard normal” distribution.
print(weights)
print(X)
```

### <a name='5'></a>  Metoda gradientu prostego  / Simple gradient method

```python
intercept=[]
coef=[]
for i in range(3000):
  gradient=(2/m)*X.T.dot(X.dot(weights)-Y)
  weights=weights-eta*gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])
print(weights)
```

```python
df=pd.DataFrame(data={'intercept':intercept,'coef':coef})
df.head()
```

### <a name='6'></a>  Wizualizacja dopasowania  / Match visualistion

```python
px.line(df,y='intercept', width=800, title='Dopsowanie Intercept')
```

```python
px.line(df,y='coef', width=800, title='Dopasowanie:coef')
```
