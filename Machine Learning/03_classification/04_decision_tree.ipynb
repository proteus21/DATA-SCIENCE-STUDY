### <a name='1'></a> Import library / Import bibliotek

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display  import Image
from sklearn.datasets import load_iris
sns.set(font_scale=1.3)
np.random.seed(42)
```

### <a name='2'></a>Loading Data / Załadowanie danych

```python
raw_data=load_iris()
all_data=raw_data.copy()

data=all_data['data']
target=all_data['target']
feature_names = [name.replace(' ', '_')[:-5] for name in all_data['feature_names']]
target_names = all_data['target_names']

print(f'Liczba próbek: {len(data)}')
print(f'Kształt danych: {data.shape}')
print(f'Nazwy zmiennych objaśniających: {feature_names}')
print(f'Nazwy kategorii kosaćca: {target_names}')
```

### <a name='3'></a> Preapare / Eksploracja danych

```python
data = df.copy()
data = data[['sepal_length', 'sepal_width', 'target']]
target = data.pop('target')

data.head()
```

```python
target.head()
```

```python
data = data.values
target = target.values.astype('int16')
```

### <a name='5'></a> Decision tree classification build / Budowa klasyfikatora drzewa decyzyjnego
max_depth=1

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth=1, random_state=42)
classifier.fit(data, target)
```

### <a name='6'></a> Decision tree classification build / Wykreslanie granic drzewa decyzyjnego


```python
from mlxtend.plotting import plot_decision_regions

colors='#f1865b,#31c30f,#64647F,#d62728,#9467bd,#8c564b,#e377c2,#7f7f7f,#bcbd22,#17becf'

acc = classifier.score(data, target)

plt.figure(figsize=(8,6))
plot_decision_regions(data, target,classifier, legend=2, colors=colors)
plt.xlabel('sepal lenght [cm]')
plt.ylabel('sepal width [cm]')
plt.title(f'Drzewo decyzyjne: max_depth=1, accuracy: {acc * 100:.2f}%')
plt.show()
```

### <a name='7'></a> Decision tree graph / Graf drzewa decyzyjnego

max_depth=1

Wyswietlenie grafu drzewa decyzyjnego

/Preview decision tree graph

```python
from six import StringIO
from sklearn.tree import export_graphviz
import pydotplus


dot_data = StringIO()
export_graphviz(classifier,
               out_file=dot_data,
               feature_names=feature_names[:2],
               class_names=target_names,
               special_characters=True,
               rounded=True,
               filled=True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('graph.png')
Image(graph.create_png(), width=300)
```

### <a name='8'></a> Decision tree graph / Budowa funkcji dla modelu drzewa decyzyjnego

```python
def make_decision_tree(max_depth=1):
  # trenowanie modelu / training model
  classifier=DecisionTreeClassifier(max_depth=max_depth, random_state=42)
  classifier.fit(data,target)

  # eksport grafu drzewa / tree graph export

  dot_data = StringIO()
  export_graphviz(classifier,
                out_file=dot_data,
                feature_names=feature_names[:2],
                class_names=target_names,
                special_characters=True,
                rounded=True,
                filled=True)
  graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
  graph.write_png('graph.png')

  # obliczenia dokładności / accurancy calculation

  acc=classifier.score(data,target)

  # wykreślenie granic decyzyjnych  /delineation of decision-making boundaries
 
  colors='#f1865b,#31c30f,#64647F,#d62728,#9467bd,#8c564b,#e377c2,#7f7f7f,#bcbd22,#17becf'
  plt.figure(figsize=(8, 6))
  ax = plot_decision_regions(data, target, classifier, legend=0, colors=colors)
  handles, labels = ax.get_legend_handles_labels()
  ax.legend(handles, ['setosa', 'versicolor', 'virginica'], framealpha=0.3)
  plt.xlabel('sepal length (cm)')
  plt.ylabel('sepal width (cm)')
  plt.title(f'Drzewo decyzyjne: max_depth={max_depth}, accuracy={acc * 100:.2f}')
  
  return Image(graph.create_png(), width=200 + max_depth * 120)
```
